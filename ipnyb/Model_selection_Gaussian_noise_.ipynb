```python
import pandas as pd
```

```python
import glob
path='/content/drive/MyDrive/DBDA_PROJECT/Gaussian_clf/'
all_files=glob.glob(path+"*.csv")
```

```python
all_files
```

```python
df = pd.DataFrame()
for file_ in all_files:
    file_df = pd.read_csv(file_,sep=',', parse_dates=[0], infer_datetime_format=True,header=None,skiprows=1 )
    file_df['model'] = file_
    df = df.append(file_df)
```

```python
df.columns =['class', 'precision', 'recall', 'f1','support','model']
```

```python
def get_filename(path):
    temp_str = path.split('/')
    return temp_str[-1]
```

```python
df["model"] = df["model"].apply(get_filename)
```

```python
# saving concat data to csv
df.to_csv("/content/drive/MyDrive/DBDA_PROJECT/Gaussian_clf/totalG.csv",index=False)
```
